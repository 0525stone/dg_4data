## Grid size 정하고 각 grid의 가로 세로 정하면, 점들을 그 그리드의 인덱스로 바꿔주는 코드
##### 참조 : 190312_3D_check.ipynb 
##### image size : 1280x720

```python
import numpy as np
import sys
```

```python
# parameter
image_x = 1280 # 얘도 파일 샘플로 하나 불러와서 width 알아내게하면 좋겠군
image_y = 720
grid_size = 80

if ((image_x%grid_size)==0):
    print('x axis ok')
else:
    print('x axis not ok')
if ((image_y%grid_size)==0):
    print('y axis ok')
else:
    print('y axis not ok')
```

```python
# reading file 1 .readlines
filename = 'L.txt'
f = open(filename, 'r')

for i in f:
    lines=f.readlines() # readline()은 줄바꿈이 문장의 단위가 됨. readlines는 문장 전체를 읽음
  #  print(lines)
#print(len(lines))

# lines를 일일이 다 잘라줘야함... readline이 이상해서
x = np.array([])
y = np.array([])

for i in range(len(lines)):
    x_,y_ = lines[i].split()
    x= np.append(x,x_)
    y= np.append(y,y_)

```

```python
# grid 화
x_g = np.array([])
y_g = np.array([])
point_g = np.array([])

for i in range(len(lines)):
    x_i = int(int(x[i])/grid_size)
    y_i = int(int(y[i])/grid_size)
   # print(x_i,'    ',y_i)
    x_g = np.append(x_g,x_i)
    y_g = np.append(y_g,y_i)

for i in range(len(lines)):
    point_g_temp = x_g[i]+y_g[i]*16
    point_g = np.append(point_g,point_g_temp)
```

```python
# file 저장
f = open("grid_L.txt",'w')
for i in range(len(point_g)):
    data=str(int(point_g[i]))+'\n'
    f.write(data)


f.close()
```
